## Clustering

```python
cluster = pyemma.coordinates.cluster_kmeans(transfeatures, k=200, max_iter=75, fixed_seed=1)
dtrajs_concatenated = np.concatenate(cluster.dtrajs)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
pyemma.plots.plot_density(
    *features_concatenated[:, :2].T, ax=ax, cbar=False, alpha=0.3)
ax.scatter(*cluster.clustercenters[:, :2].T, s=5, c='C1')
ax.set_xlabel('Trp $\chi1$')
ax.set_ylabel('Trp $\chi2$')
fig.tight_layout()
```

## ITS

```python
fig, ax = plt.subplots(figsize=(10, 8))
its = pyemma.msm.its(cluster.dtrajs, lags=[10, 100, 200, 300, 400, 1000, 2000, 3000], nits=3, errors='bayes')
pyemma.plots.plot_implied_timescales(its, units='ns', dt=1/50);
ax.set_xlabel('lag time (ns)', fontsize=25)
ax.set_ylabel('timescale /ns', fontsize=25)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
its = pyemma.msm.its(cluster.dtrajs, lags=[10, 100, 200, 300, 400, 500, 800, 1000], nits=3, errors='bayes')
pyemma.plots.plot_implied_timescales(its, units='ns', dt=1/50);
ax.set_xlabel('lag time (ns)', fontsize=25)
ax.set_ylabel('timescale /ns', fontsize=25)
```

## MSM

```python
lag = 6*50
msm = pyemma.msm.bayesian_markov_model(cluster.dtrajs, lag=lag)
print('fraction of states used = {:.2f}'.format(msm.active_state_fraction))
print('fraction of counts used = {:.2f}'.format(msm.active_count_fraction))
```

## Free Energy Surface

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharex=True, sharey=True)
pyemma.plots.plot_contour(
    *features_concatenated[:, :2].T,
    msm.pi[np.concatenate(msm.dtrajs_active)],
    ax=axes[0],
    mask=True,
    cbar_label='stationary distribution')
pyemma.plots.plot_free_energy(
    *features_concatenated[:, :2].T,
    weights=np.concatenate(msm.trajectory_weights()),
    ax=axes[1],
    legacy=False)
for ax in axes.flat:
    ax.set_xlabel('Trp $\chi1$')
axes[0].set_ylabel('Trp $\chi2$')
axes[0].set_title('Stationary distribution', fontweight='bold')
axes[1].set_title('Reweighted free energy surface', fontweight='bold')
fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
pyemma.plots.plot_free_energy(*features_concatenated[:, :2].T, weights=np.concatenate(msm.trajectory_weights()), ax=ax, legacy=False, alpha=0.8)
ax.set_xlabel('Trp $\chi1$ (radian)', fontsize=30)
ax.set_ylabel('Trp $\chi2$ (radian)', fontsize=30)
ax.set_xlim(-np.pi, np.pi)
ax.set_ylim(-np.pi, np.pi)
ax.set_aspect('equal')

fig.tight_layout()
```

```python
metastable_traj = msm.metastable_assignments[np.concatenate(msm.dtrajs_active)]

fig, ax = plt.subplots(figsize=(10, 8))
_, _, misc = pyemma.plots.plot_state_map(
    *features_concatenated[:, :2].T, metastable_traj, ax=ax)
ax.set_xlabel('Trp $\chi1$ (radian)', fontsize=30)
ax.set_ylabel('Trp $\chi2$ (radian)', fontsize=30)
misc['cbar'].set_ticklabels([r'$\mathcal{S}_%d$' % (i + 1)
                             for i in range(nstates)])
fig.tight_layout()
```

## MFPT

```python
from itertools import product

mfpt = np.zeros((nstates, nstates))
for i, j in product(range(nstates), repeat=2):
    mfpt[i, j] = msm.mfpt(
        msm.metastable_sets[i],
        msm.metastable_sets[j])

from pandas import DataFrame
print('MFPT (ns)')
DataFrame(np.round(mfpt /50, decimals=2), index=range(1, nstates + 1), columns=range(1, nstates + 1))
```
